```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\DELL\\Downloads\\starlink_tle_dataset_csv_type.csv")
print("Shape of dataset:", df.shape)
df.head()
```

```python
print("Basic Information about Dataset:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

```python
#inclination distribution 
plt.figure(figsize=(6,4))
plt.hist(df["Inclination_deg"], bins=25, color="red", edgecolor="black")
plt.xlabel("Inclination (degrees)")
plt.ylabel("Count")
plt.title("Distribution of Satellite Inclinations")
plt.show()
```

```python
#eccentricity distribution
plt.figure(figsize=(6,4))
plt.boxplot(df["Eccentricity"], vert=False, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.xlabel("Eccentricity")
plt.title("Eccentricity Spread (Boxplot)")
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(8,6))
corr = df.drop(columns=["Satellite_Name","Epoch"]).corr()
sns.heatmap(corr, annot=True, cmap="pink", fmt=".2f")
plt.title("Correlation Heatmap of Orbital Parameters")
plt.show()
```

```python
#orbit period and semi major axiz 
plt.figure(figsize=(6,4))
plt.scatter(df["Semi_Major_Axis_km"], df["Orbital_Period_min"], s=10, alpha=0.6)
plt.xlabel("Semi-Major Axis (km)")
plt.ylabel("Orbital Period (min)")
plt.title("Semi-Major Axis vs Orbital Period")
plt.show()
```
